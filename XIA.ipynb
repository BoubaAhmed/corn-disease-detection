```python
# ========== SIMPLE XAI TEST FOR LOCAL IMAGE (FIXED) ==========

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

# Load your trained model
model_path = "corn_leaf_disease_model_XAI.keras"  # Update this path
model = load_model(model_path)
print("✅ Model loaded successfully!")

# Define class names (update based on your model)
CLASS_NAMES = ["Blight", "Common_Rust", "Gray_Leaf_Spot", "Healthy"]

def preprocess_single_image(img_path, target_size=(384, 384)):
    """Load and preprocess a single image"""
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array, img

def get_last_conv_layer(model):
    """Find the last convolutional layer"""
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    
    # If no Conv2D found, search in nested layers
    for layer in reversed(model.layers):
        if hasattr(layer, 'layers'):
            for sub_layer in reversed(layer.layers):
                if isinstance(sub_layer, tf.keras.layers.Conv2D):
                    return sub_layer.name
    
    raise ValueError("No convolutional layer found!")

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """Generate Grad-CAM heatmap"""
    try:
        grad_model = tf.keras.models.Model(
            inputs=model.inputs,
            outputs=[model.get_layer(last_conv_layer_name).output, model.output]
        )

        with tf.GradientTape() as tape:
            conv_outputs, predictions = grad_model(img_array)
            if isinstance(conv_outputs, (list, tuple)):
                conv_outputs = conv_outputs[0]
            if isinstance(predictions, (list, tuple)):
                predictions = predictions[0]

            if pred_index is None:
                pred_index = tf.argmax(predictions[0])
            class_channel = predictions[:, pred_index]

        grads = tape.gradient(class_channel, conv_outputs)
        pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
        conv_outputs = conv_outputs[0]

        heatmap = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)
        heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)

        # Always cast pred_index to int
        return heatmap.numpy(), int(pred_index)

    except Exception as e:
        print(f"❌ Error in Grad-CAM: {e}")
        return None, None
def explain_single_image(model, image_path, class_names):
    """Complete XAI analysis for a single image"""
    print(f"🔍 Analyzing: {os.path.basename(image_path)}")
    
    try:
        # 1. Preprocess image
        img_array, original_img = preprocess_single_image(image_path)
        
        # 2. Make prediction
        predictions = model.predict(img_array, verbose=0)
        pred_class = np.argmax(predictions[0])
        confidence = np.max(predictions[0])
        
        print(f"🎯 Prediction: {class_names[pred_class]}")
        print(f"📊 Confidence: {confidence:.2%}")
        
        # 3. Find conv layer and generate heatmap
        last_conv_layer_name = get_last_conv_layer(model)
        print(f"📋 Using layer: {last_conv_layer_name}")
        
        heatmap, _ = make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_class)
        
        if heatmap is None:
            print("❌ Could not generate heatmap")
            return None
        
        # 4. Prepare images for display
        original_img_display = (img_array[0] * 255).astype(np.uint8)
        
        # Resize heatmap and create overlay
        heatmap_resized = cv2.resize(heatmap, (original_img.size[0], original_img.size[1]))
        heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
        
        # Convert original image to BGR for overlay (OpenCV uses BGR)
        original_img_bgr = cv2.cvtColor(np.array(original_img), cv2.COLOR_RGB2BGR)
        superimposed_img = cv2.addWeighted(original_img_bgr, 0.6, heatmap_colored, 0.4, 0)
        
        # 5. Create visualization
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        
        # Original image
        axes[0, 0].imshow(original_img)
        axes[0, 0].set_title('Original Image', fontweight='bold')
        axes[0, 0].axis('off')
        
        # Prediction info
        axes[0, 1].text(0.1, 0.9, 'PREDICTION RESULTS', fontsize=14, fontweight='bold', transform=axes[0, 1].transAxes)
        axes[0, 1].text(0.1, 0.7, f'Class: {class_names[pred_class]}', fontsize=12, transform=axes[0, 1].transAxes)
        axes[0, 1].text(0.1, 0.6, f'Confidence: {confidence:.2%}', fontsize=12, transform=axes[0, 1].transAxes)
        axes[0, 1].text(0.1, 0.4, 'All Probabilities:', fontsize=11, fontweight='bold', transform=axes[0, 1].transAxes)
        
        y_pos = 0.3
        for i, prob in enumerate(predictions[0]):
            color = 'red' if i == pred_class else 'black'
            axes[0, 1].text(0.1, y_pos, f'{class_names[i]}: {prob:.2%}', fontsize=10, color=color, transform=axes[0, 1].transAxes)
            y_pos -= 0.08
        
        axes[0, 1].set_xlim(0, 1)
        axes[0, 1].set_ylim(0, 1)
        axes[0, 1].axis('off')
        
        # Grad-CAM result
        axes[1, 0].imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
        axes[1, 0].set_title('Grad-CAM Overlay', fontweight='bold')
        axes[1, 0].axis('off')
        
        # Heatmap alone
        axes[1, 1].imshow(heatmap, cmap='jet')
        axes[1, 1].set_title('Heatmap Only', fontweight='bold')
        axes[1, 1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        return {
            'predicted_class': pred_class,
            'predicted_class_name': class_names[pred_class],
            'confidence': confidence,
            'all_predictions': predictions[0]
        }
    
    except Exception as e:
        print(f"❌ Error during analysis: {e}")
        import traceback
        traceback.print_exc()
        return None

# ========== USAGE ==========

# Test with your image
image_path = "Corn_Blight (1075).jpeg"  # Update this path to your test image

if os.path.exists(image_path):
    result = explain_single_image(model, image_path, CLASS_NAMES)
    if result:
        print(f"\n✅ Analysis completed!")
        print(f"Result: {result['predicted_class_name']} ({result['confidence']:.2%})")
else:
    print(f"❌ Image not found: {image_path}")
    print("Please update the 'image_path' variable with your image path")
```

```python
# ========== XAI AVEC LÉGENDE ET ANALYSE DÉTAILLÉE DES HEATMAPS ==========

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

# Charger le modèle
model_path = "corn_leaf_disease_model_XAI.keras"
model = load_model(model_path)
print("✅ Modèle chargé avec succès!")

# Noms des classes
CLASS_NAMES = ["Blight", "Common_Rust", "Gray_Leaf_Spot", "Healthy"]

def preprocess_single_image(img_path, target_size=(384, 384)):
    """Charger et prétraiter une seule image"""
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array, img

def get_last_conv_layer(model):
    """Trouver la dernière couche convolutionnelle"""
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    
    # Si aucune Conv2D trouvée, chercher dans les couches imbriquées
    for layer in reversed(model.layers):
        if hasattr(layer, 'layers'):
            for sub_layer in reversed(layer.layers):
                if isinstance(sub_layer, tf.keras.layers.Conv2D):
                    return sub_layer.name
    
    raise ValueError("Aucune couche convolutionnelle trouvée!")

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """Générer la heatmap Grad-CAM"""
    try:
        grad_model = tf.keras.models.Model(
            inputs=model.inputs,
            outputs=[model.get_layer(last_conv_layer_name).output, model.output]
        )

        with tf.GradientTape() as tape:
            conv_outputs, predictions = grad_model(img_array)
            if isinstance(conv_outputs, (list, tuple)):
                conv_outputs = conv_outputs[0]
            if isinstance(predictions, (list, tuple)):
                predictions = predictions[0]

            if pred_index is None:
                pred_index = tf.argmax(predictions[0])
            class_channel = predictions[:, pred_index]

        grads = tape.gradient(class_channel, conv_outputs)
        pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
        conv_outputs = conv_outputs[0]

        heatmap = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)
        heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)

        return heatmap.numpy(), int(pred_index)

    except Exception as e:
        print(f"❌ Erreur dans Grad-CAM: {e}")
        return None, None

def analyze_heatmap_distribution(heatmap, original_img):
    """Analyser la distribution de la heatmap pour une interprétation détaillée"""
    # Calculer les statistiques de la heatmap
    heatmap_flat = heatmap.flatten()
    
    # Seuils pour différentes intensités
    low_threshold = 0.1
    medium_threshold = 0.3
    high_threshold = 0.6
    
    # Calculer les pourcentages
    low_activation = np.sum(heatmap_flat > low_threshold) / len(heatmap_flat) * 100
    medium_activation = np.sum(heatmap_flat > medium_threshold) / len(heatmap_flat) * 100
    high_activation = np.sum(heatmap_flat > high_threshold) / len(heatmap_flat) * 100
    
    # Trouver les régions les plus activées
    max_activation_coords = np.unravel_index(np.argmax(heatmap), heatmap.shape)
    
    # Calculer la couverture globale
    coverage = np.sum(heatmap_flat > 0.05) / len(heatmap_flat) * 100
    
    return {
        'max_activation': np.max(heatmap),
        'mean_activation': np.mean(heatmap),
        'coverage_percentage': coverage,
        'low_activation_percentage': low_activation,
        'medium_activation_percentage': medium_activation,
        'high_activation_percentage': high_activation,
        'max_activation_coords': max_activation_coords,
        'activation_variance': np.var(heatmap)
    }

def explain_single_image_detailed(model, image_path, class_names):
    """Analyse XAI complète avec légende détaillée et interprétation"""
    print(f"🔍 Analyse détaillée de: {os.path.basename(image_path)}")
    
    try:
        # 1. Prétraiter l'image
        img_array, original_img = preprocess_single_image(image_path)
        
        # 2. Faire la prédiction
        predictions = model.predict(img_array, verbose=0)
        pred_class = np.argmax(predictions[0])
        confidence = np.max(predictions[0])
        
        print(f"🎯 Prédiction: {class_names[pred_class]}")
        print(f"📊 Confiance: {confidence:.2%}")
        
        # 3. Trouver la couche conv et générer la heatmap
        last_conv_layer_name = get_last_conv_layer(model)
        print(f"📋 Utilisation de la couche: {last_conv_layer_name}")
        
        heatmap, _ = make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_class)
        
        if heatmap is None:
            print("❌ Impossible de générer la heatmap")
            return None
        
        # 4. Analyser la distribution de la heatmap
        heatmap_analysis = analyze_heatmap_distribution(heatmap, original_img)
        
        # 5. Préparer les images pour l'affichage
        original_img_np = np.array(original_img)
        
        # Redimensionner la heatmap et créer l'overlay
        heatmap_resized = cv2.resize(heatmap, (original_img.size[0], original_img.size[1]))
        heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
        
        # Convertir l'image originale en BGR pour l'overlay
        original_img_bgr = cv2.cvtColor(original_img_np, cv2.COLOR_RGB2BGR)
        superimposed_img = cv2.addWeighted(original_img_bgr, 0.6, heatmap_colored, 0.4, 0)
        
        # 6. Créer la visualisation améliorée avec légende
        fig = plt.figure(figsize=(20, 12))
        
        # Configuration de la grille
        gs = plt.GridSpec(3, 3, figure=fig)
        
        # Image originale
        ax1 = fig.add_subplot(gs[0, 0])
        ax1.imshow(original_img)
        ax1.set_title('🖼️ Image Originale', fontweight='bold', fontsize=14, pad=20)
        ax1.axis('off')
        
        # Résultats de prédiction
        ax2 = fig.add_subplot(gs[0, 1])
        ax2.text(0.1, 0.95, '📊 RÉSULTATS DE PRÉDICTION', fontsize=16, fontweight='bold', 
                transform=ax2.transAxes, color='darkblue')
        ax2.text(0.1, 0.85, f'🔍 Classe: {class_names[pred_class]}', fontsize=14, 
                transform=ax2.transAxes, color='red' if class_names[pred_class] != 'Healthy' else 'green')
        ax2.text(0.1, 0.80, f'🎯 Confiance: {confidence:.2%}', fontsize=14, 
                transform=ax2.transAxes)
        
        # Probabilités détaillées
        ax2.text(0.1, 0.70, '📈 Probabilités détaillées:', fontsize=12, fontweight='bold', 
                transform=ax2.transAxes)
        
        y_pos = 0.65
        for i, prob in enumerate(predictions[0]):
            color = 'red' if i == pred_class else 'black'
            marker = '✅' if i == pred_class else '  '
            ax2.text(0.1, y_pos, f'{marker} {class_names[i]}: {prob:.2%}', 
                    fontsize=11, color=color, transform=ax2.transAxes)
            y_pos -= 0.06
        
        ax2.set_xlim(0, 1)
        ax2.set_ylim(0, 1)
        ax2.axis('off')
        
        # Overlay Grad-CAM
        ax3 = fig.add_subplot(gs[1, 0])
        ax3.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
        ax3.set_title('🎨 Overlay Grad-CAM', fontweight='bold', fontsize=14, pad=20)
        ax3.axis('off')
        
        # Heatmap seule avec légende
        ax4 = fig.add_subplot(gs[1, 1])
        im = ax4.imshow(heatmap, cmap='jet')
        ax4.set_title('🔥 Heatmap Seule', fontweight='bold', fontsize=14, pad=20)
        ax4.axis('off')
        
        # Ajouter la légende de couleur
        cbar = plt.colorbar(im, ax=ax4, fraction=0.046, pad=0.04)
        cbar.set_label("Intensité d'activation", rotation=270, labelpad=20)
        
        # Analyse détaillée de la heatmap
        ax5 = fig.add_subplot(gs[0, 2])
        ax5.text(0.1, 0.95, '🔬 ANALYSE DÉTAILLÉE DE LA HEATMAP', fontsize=16, 
                fontweight='bold', transform=ax5.transAxes, color='darkgreen')
        
        analysis_text = [
            f"📏 Activation maximale: {heatmap_analysis['max_activation']:.3f}",
            f"📊 Activation moyenne: {heatmap_analysis['mean_activation']:.3f}",
            f"📐 Variance d'activation: {heatmap_analysis['activation_variance']:.5f}",
            "",
            f"🎯 Couverture de la heatmap: {heatmap_analysis['coverage_percentage']:.1f}%",
            f"🔵 Faible activation (>0.1): {heatmap_analysis['low_activation_percentage']:.1f}%",
            f"🟡 Activation moyenne (>0.3): {heatmap_analysis['medium_activation_percentage']:.1f}%",
            f"🔴 Forte activation (>0.6): {heatmap_analysis['high_activation_percentage']:.1f}%",
            "",
            f"📍 Point d'activation max: {heatmap_analysis['max_activation_coords']}"
        ]
        
        y_pos_text = 0.85
        for line in analysis_text:
            ax5.text(0.1, y_pos_text, line, fontsize=11, transform=ax5.transAxes,
                   verticalalignment='top')
            y_pos_text -= 0.06
        
        ax5.set_xlim(0, 1)
        ax5.set_ylim(0, 1)
        ax5.axis('off')
        
        # Interprétation des résultats
        ax6 = fig.add_subplot(gs[1, 2])
        ax6.text(0.1, 0.95, '💡 INTERPRÉTATION DES RÉSULTATS', fontsize=16, 
                fontweight='bold', transform=ax6.transAxes, color='purple')
        
        # Générer l'interprétation basée sur l'analyse
        interpretation = generate_interpretation(heatmap_analysis, class_names[pred_class], confidence)
        
        y_pos_interp = 0.85
        for line in interpretation:
            ax6.text(0.1, y_pos_interp, line, fontsize=11, transform=ax6.transAxes,
                   verticalalignment='top', style='italic')
            y_pos_interp -= 0.07
        
        ax6.set_xlim(0, 1)
        ax6.set_ylim(0, 1)
        ax6.axis('off')
        
        # Légende explicative
        ax7 = fig.add_subplot(gs[2, :])
        legend_text = [
            "📖 LÉGENDE - COMMENT INTERPRÉTER LA HEATMAP:",
            "• 🔵 Zones BLEUES: Faible activation - Le modèle accorde peu d'attention à ces régions",
            "• 🟢 Zones VERTES: Activation moyenne - Régions modérément importantes pour la décision", 
            "• 🟡 Zones JAUNES: Activation forte - Régions importantes pour la classification",
            "• 🔴 Zones ROUGES: Activation maximale - Régions critiques pour la décision du modèle",
            "",
            "💡 CONSEILS D'INTERPRÉTATION:",
            "• Une heatmap FOCALISÉE (forte activation sur petites zones) indique une bonne spécificité",
            "• Une heatmap DIFFUSE peut indiquer que le modèle utilise des features générales",
            "• Les points d'activation MAXIMALE montrent les régions les plus déterminantes"
        ]
        
        y_pos_legend = 0.95
        for i, line in enumerate(legend_text):
            weight = 'bold' if i == 0 else 'normal'
            color = 'darkblue' if i == 0 else 'black'
            ax7.text(0.02, y_pos_legend, line, fontsize=10, transform=ax7.transAxes,
                   verticalalignment='top', weight=weight, color=color)
            y_pos_legend -= 0.10 if i == 0 else 0.07
        
        ax7.set_xlim(0, 1)
        ax7.set_ylim(0, 1)
        ax7.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # 7. Afficher le résumé dans la console
        print(f"\n{'='*60}")
        print("📊 RÉSUMÉ DE L'ANALYSE XAI")
        print(f"{'='*60}")
        print(f"🔍 Prédiction: {class_names[pred_class]} ({confidence:.2%})")
        print(f"📏 Couverture de la heatmap: {heatmap_analysis['coverage_percentage']:.1f}%")
        print(f"🎯 Activation maximale: {heatmap_analysis['max_activation']:.3f}")
        print(f"📊 Focus du modèle: {'ÉLEVÉ' if heatmap_analysis['high_activation_percentage'] < 20 else 'MODÉRÉ'}")
        print(f"💡 Interprétation: {interpretation[0]}")
        
        return {
            'predicted_class': pred_class,
            'predicted_class_name': class_names[pred_class],
            'confidence': confidence,
            'all_predictions': predictions[0],
            'heatmap_analysis': heatmap_analysis
        }
    
    except Exception as e:
        print(f"❌ Erreur pendant l'analyse: {e}")
        import traceback
        traceback.print_exc()
        return None

def generate_interpretation(heatmap_analysis, predicted_class, confidence):
    """Générer une interprétation textuelle basée sur l'analyse de la heatmap"""
    interpretations = []
    
    # Basé sur la couverture
    coverage = heatmap_analysis['coverage_percentage']
    if coverage < 15:
        interpretations.append("• Le modèle est très FOCALISÉ sur des régions spécifiques")
    elif coverage < 40:
        interpretations.append("• Le modèle utilise une combinaison de régions locales et globales")
    else:
        interpretations.append("• Le modèle a une vision GLOBALE de l'image")
    
    # Basé sur l'activation forte
    high_activation = heatmap_analysis['high_activation_percentage']
    if high_activation < 10:
        interpretations.append("• Forte SPÉCIALISATION: peu de régions très importantes")
    elif high_activation < 25:
        interpretations.append("• Équilibre entre spécificité et contexte global")
    else:
        interpretations.append("• Multiples régions importantes pour la décision")
    
    # Basé sur la variance
    variance = heatmap_analysis['activation_variance']
    if variance > 0.02:
        interpretations.append("• Fort CONTRASTE d'activation entre régions")
    else:
        interpretations.append("• Activation relativement UNIFORME across l'image")
    
    # Basé sur la confiance
    if confidence > 0.95:
        interpretations.append("• Prédiction TRÈS CONFIANTE avec features clairs")
    elif confidence > 0.8:
        interpretations.append("• Prédiction CONFIANTE avec bonnes indications")
    else:
        interpretations.append("• Prédiction MODÉRÉE, pourrait bénéficier de plus de données")
    
    # Ajouter une conclusion
    if predicted_class != "Healthy" and high_activation < 20:
        interpretations.append("• Le modèle identifie clairement les symptômes de maladie")
    elif predicted_class == "Healthy":
        interpretations.append("• Pattern caractéristique de feuille saine détecté")
    
    return interpretations

# ========== UTILISATION ==========

# Tester avec votre image
image_path = "Corn_Blight (1084).jpeg"  # Mettez à jour avec le chemin de votre image

if os.path.exists(image_path):
    result = explain_single_image_detailed(model, image_path, CLASS_NAMES)
    if result:
        print(f"\n✅ Analyse complétée avec succès!")
        print(f"🎯 Résultat final: {result['predicted_class_name']} ({result['confidence']:.2%})")
else:
    print(f"❌ Image non trouvée: {image_path}")
    print("Veuillez mettre à jour la variable 'image_path' avec votre chemin d'image")
```
